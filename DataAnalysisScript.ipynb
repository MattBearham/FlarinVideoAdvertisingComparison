```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
import collections
import string
import matplotlib.pyplot as plt
```

```python
feedback = pd.read_csv("feedback.csv")

#How many results you want to see
n = 5
```

```python
def preprocess_data(column):
    column_joined = ' '.join(column)
    column_tokens = word_tokenize(column_joined)
    column_clean = [word.lower() for word in column_tokens if word not in string.punctuation]
    column_no_sw = [word for word in column_clean if not word in stopwords.words()]
    return column_no_sw
```

```python
def generate_data(preprocessed_data):
    data_wordcount = {}
    for word in preprocessed_data:
        if word not in data_wordcount:
            data_wordcount[word] = 1
        else:
            data_wordcount[word] += 1

    data_word_counter = collections.Counter(data_wordcount)

    data_words = []
    data_counts = []

    for word, count in data_word_counter.most_common(n):
        #print(word, ": ", count)
        data_words.append(word)
        data_counts.append(count)
    
    return data_words,data_counts
```

```python
#Video A preprocessing

a_message = feedback["a_message"]
a_like = feedback["a_like"]
a_dislike = feedback["a_dislike"]

a_message_tokens_no_sw = preprocess_data(a_message)
a_like_tokens_no_sw = preprocess_data(a_like)
a_dislike_tokens_no_sw = preprocess_data(a_dislike)
```

```python
#Video A data generation

a_message_words,a_message_counts = generate_data(a_message_tokens_no_sw)
a_like_words,a_like_counts = generate_data(a_like_tokens_no_sw)
a_dislike_words,a_dislike_counts = generate_data(a_dislike_tokens_no_sw)
```
